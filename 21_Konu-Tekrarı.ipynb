```python
def not_hesap():
    vize = int(input("Vize notunuzu giriniz: "))
    final = int(input("Final notunuzu giriniz: "))
    
    not_ort = (vize*0.4) + (final*0.6)
    
    if not_ort >= 90:
        print(f'''
        Tebrikler dersi geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: AA
        ''')
        
    elif not_ort >= 85:
        print(f'''
        Tebrikler dersi geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: BA
        ''')
    
    elif not_ort >= 80:
        print(f'''
        Tebrikler dersi geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: BB
        ''')
    
    elif not_ort >= 70:
        print(f'''
        Tebrikler dersi geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: CB
        ''')
    
    elif not_ort >= 60:
        print(f'''
        Tebrikler dersi geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: CC
        ''')
    
    elif not_ort >= 55:
        print(f'''
        Dersi koşullu geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: DC
        ''')
    
    elif not_ort >= 50:
        print(f'''
        Dersi koşullu geçtiniz, not ortalamanız: {not_ort}
        Harf notunuz: DD
        ''')
    
    else:
        print(f'''
        Dersten kaldınız, not ortalamanız: {not_ort}
        Harf notunuz: FF
        ''')
```

```python
while True:
    not_hesap()
```

```python
def topla(*args):
    toplam = 0
    for i in args:
        toplam += i
    return toplam
```

```python
topla(1,2,3)
```

```python
#list yapısını tek bir argüman olarak alacağı için fonksiyonumuz çalışmaz ve hata alırız
topla([1,2,3,4])
```

```python
# Fibonacci Recursion

# 0 1 1 2 3 5 8 13

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))

x = int(input("Fibonacci hesaplamak için değer giriniz: "))

for i in range(x):
    print(fibonacci(i))


#fibonacci(0) = 0
#fibonacci(1) = 1
#fibonacci(2) = fibonacci(1) + fibonacci(0) = 1+0 = 1
#fibonacci(3) = fibonacci(2) + fibonacci(1) = 1+1 = 2
#fibonacci(4) = fibonacci(3) + fibonacci(2) = 2+1 = 3
#fibonacci(5) = fibonacci(4) + fibonacci(3) = 3+2 = 5
#fibonacci(6) = fibonacci(5) + fibonacci(4) = 5+3 = 8
#fibonacci(7) = fibonacci(6) + fibonacci(5) = 8+5 = 13
```
